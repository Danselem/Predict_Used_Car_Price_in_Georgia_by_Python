```python
url = 'https://www.cars.com/for-sale/searchresults.action/?page=2&perPage=20&rd=50&searchSource=PAGINATION&showMore=true&sort=relevance&stkTypId=28881&zc=30265'
```

```python
sauce = urllib.request.urlopen(url).read()
soup = bs.BeautifulSoup(sauce, 'lxml')
```

```python
specificSoup = soup.find_all('div', class_='listing-row__details')
```
